Un  sitio  de  E books  tiene  información  sobre  los  reviews  que  los  usuarios hacen  de  sus  libros  en  un  DataFrame  con  formato  (user_id,  book_id,  rating, timestamp). 

Por otro lado tenemos información en otro DataFrame que bajamos de  GoodReads:  (book_id,  book_name,  avg_rating).  Podemos  suponer  que  los Ids de los libros son compatibles. 

Se pide usar Python Pandas para:

1. Obtener un DataFrame que indique el TOP5 de Ebooks en el sitio de Ebooks. (Para este punto se puede ignorar el segundo DataFrame).
2. Obtener un DataFrame que indique qué libros tienen una diferencia de rating promedio mayor al 20% entre el sitio de Ebooks y GoodReads.

```python
# importacion general de librerias y de visualizacion (matplotlib y seaborn)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random as rd
%matplotlib inline

plt.style.use('default') # haciendo los graficos un poco mas bonitos en matplotlib
#plt.rcParams['figure.figsize'] = (20, 10)

sns.set(style="whitegrid") # seteando tipo de grid en seaborn

pd.options.display.float_format = '{:20,.2f}'.format # suprimimos la notacion cientifica en los outputs

import warnings
warnings.filterwarnings('ignore')
```

```python
ebooks = pd.DataFrame({
    'user_id': rd.choices(range(1,50),k=80),
    'book_id': rd.choices(range(1,16),k=80),
    'rating': rd.choices(range(1,6),k=80),
    'timestamp':pd.date_range(start='2021-04-01',freq='H',periods=80)
})
ebooks
```

```python
goodread = pd.DataFrame({
    'book_id':range(1,16),
    'book_name': ['libro1','libro2','libro3','libro4','libro5','libro6','libro7','libro8','libro9','libro10','libro11','libro12','libro13','libro14','libro15'],
    'avg_rating': rd.choices(range(1,6),k=15)
})
goodread
```

## ejercicio 1
1. Obtener un DataFrame que indique el TOP5 de Ebooks en el sitio de Ebooks. (Para este punto se puede ignorar el segundo DataFrame).


```python
libros_por_id = ebooks.groupby('book_id').mean()
libros_por_id.nlargest(5,'rating').reset_index()[['book_id','rating']]
```

```python
libros_por_id = libros_por_id.rename(columns={'rating':'rating_ebook'})
goodread = goodread.rename(columns={'avg_rating':'rating_goodread'})
libros_ratings = pd.merge(libros_por_id,goodread,on='book_id',how='inner')[['book_name','rating_ebook','rating_goodread']]
libros_ratings['difference_perc'] = (libros_ratings[['rating_ebook','rating_goodread']].max(axis=1)/libros_ratings[['rating_ebook','rating_goodread']].min(axis=1))-1
libros_ratings[libros_ratings['difference_perc']>0.2]
```
