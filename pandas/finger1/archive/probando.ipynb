```python
#Creamos el dataframe
import pandas as pd
df = pd.read_csv("country_vaccinations.csv")
df.tail()
```

```python
df.groupby('country').agg({'people_vaccinated':'max','people_fully_vaccinated':'max'}).sort_values(by='people_vaccinated',ascending=False)
```

```python
grouped = df.groupby('country').agg({'people_vaccinated':'max'})
grouped
```

```python
sorted_by_people_vaccinated = grouped.sort_values(by='people_vaccinated',ascending=False)
sorted_by_people_vaccinated.index.get_loc('Germany')
```
