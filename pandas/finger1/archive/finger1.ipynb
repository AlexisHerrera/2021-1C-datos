```python
#Creamos el dataframe
import pandas as pd
df = pd.read_csv("country_vaccinations_by_manufacturer.csv")
```

```python
#Agrupo por país
vacs_by_country = df.groupby('location').agg(list)
vacs_by_country.rename(columns={'total_vaccinations':'vaccinations_per_day'},inplace=True)
vacs_by_country
```

```python
#Creo la columna total de vacunados y ordeno
vacs_by_country['total_vaccinations'] = vacs_by_country['vaccinations_per_day'].apply(sum)
vacs_by_country
```

```python
#Muestro los datos que nos interesan ordenados
vacs_by_country_ordered = vacs_by_country[['total_vaccinations']].reset_index().sort_values(by='total_vaccinations',ascending=False)
vacs_by_country_ordered
```
